```python
from sklearn.metrics.pairwise import cosine_similarity

# Step 1: compute similarity matrix
similarity_scores = cosine_similarity(pt)

# Step 2: recommendation function
def recommend(book_name):
    # find index of the book
    index = np.where(pt.index == book_name)[0][0]
    
    # get similarity vector
    distances = similarity_scores[index]
    
    # get top 5 similar books (excluding itself)
    similar_items = sorted(list(enumerate(distances)), 
                           key=lambda x: x[1], 
                           reverse=True)[1:6]
    
    recommendations = []
    for i in similar_items:
        recommendations.append(pt.index[i[0]])

    return recommendations
```

```python
def recommend(book_name):
    # check if book exists in index
    if book_name not in pt.index:
        return f"❌ Book '{book_name}' not found in dataset!"

    # find index of the book
    index = np.where(pt.index == book_name)[0][0]

    # get similarity vector
    distances = similarity_scores[index]

    # get top 5 similar books (excluding itself)
    similar_items = sorted(
        list(enumerate(distances)),
        key=lambda x: x[1],
        reverse=True
    )[1:6]

    recommendations = []
    for i in similar_items:
        recommendations.append(pt.index[i[0]])

    return recommendations
```

```python
book = input("Enter a book name: ")

result = recommend(book)
print("\nRecommended Books:\n")

if isinstance(result, list):
    for r in result:
        print("📘", r)
else:
    print(result)
```
